```python
#Usado para Realizar Calculos do DIFAL

import PySimpleGUI as sg

sg.theme('Topanga')

X = sg.popup_get_text('Quantidade de Aliquotas')
M = int(X)

NN = []

for i in range(M):


    layout = [

    [sg.Text('Coloque o Valor da Nota (ao inves de , use .)')],
    [sg.InputText('', key='v_n')],

    [sg.Text('Coloque a Aliquota da Nota (ao inves de , use .)')],
    [sg.InputText('', key='a_n')],
    
    [sg.Button('OK'), sg.Button('Cancelar')]
    ]

    window = sg.Window('Entrada de Dados', layout)

    while True:
        event, values = window.read()
        
        if event == sg.WINDOW_CLOSED or event == 'Cancelar':
            break
        
        if event == 'OK':
            v_n = values['v_n']
            a_n = values['a_n']
    
            window.close()
            
    window.close()

    dv_n = float(v_n)

    da_n = float(a_n)

    if da_n == 4:

        d1 = (dv_n * 0.0787) #7,87% 
        dd1 = '{0:.2f}'.format(d1)

        dd1 = float(dd1)
        
        NN.append(dd1)
        
        print('Aliquota {} (R${} * 7,87% = R${})'.format(a_n, dv_n, dd1))

    if da_n >= 5:

        d2 = (dv_n * 0.0449) #4,49%
        dd2 = '{0:.2f}'.format(d2)

        dd2 = float(dd2)

        NN.append(dd2)

        print('Aliquota {} (R${} * 4,49% = R${})'.format(a_n, dv_n, dd2))

    if da_n == 0:

        d3 = (dv_n * 0.0449) #4,49%
        dd3 = '{0:.2f}'.format(d3)

        dd3 = float(dd3)

        NN.append(dd3)

        print('Aliquota {} (R${} * 4,49% = R${})'.format(a_n, dv_n, dd3))

soma = sum(NN)
soma = '{0:.2f}'.format(soma)
print('Total de {}'.format(soma))
```
