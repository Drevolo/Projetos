```python
#Tela interativa usada para Calcular Impostos

import PySimpleGUI as sg

sg.theme("Topanga")

layout = [
    [sg.Text('Coloque o Valor (ao inves de , use .)')], 
    [sg.InputText(key='valor')],
    [sg.Text('PIS R$'), sg.Text('', key="vpis")],
    [sg.Text('CONF R$'), sg.Text('', key="vconf")],
    [sg.Text('IR R$'), sg.Text('', key="vir")],
    [sg.Text('CSLL R$'), sg.Text('', key="vcsll")],
    [sg.Text('Total R$'), sg.Text('', key="ftotal")]
    [sg.Button('Calcular'),sg.Button('Fechar')]
]

janela = sg.Window('Tela', layout)

while True:

    eventos, valores = janela.read()

    if eventos == sg.WINDOW_CLOSED or eventos == 'Fechar':
        break

    if eventos == 'Calcular':

        valor = valores['valor']
        decimal = float(valor)
        ddecimal = '{0:.2f}'.format(decimal)

        #PIS
        pis = (decimal * 0.0065)
        ppis = '{0:.2f}'.format(pis)
        ppis = float(ppis)

        pis2 = (decimal * 0.0065)
        ppis2 = '{0:.2f}'.format(pis2)
        ppis2 = float(ppis2)

        pis3 = (ppis2 - ppis)
        ppis3 = '{0:.2f}'.format(pis3)

        janela["vpis"].update(ppis3)

        #CONF
        conf = (decimal * 0.03)
        cconf = '{0:.2f}'.format(conf)
        cconf = float(cconf)

        conf2 = (decimal * 0.03)
        cconf2 = '{0:.2f}'.format(conf2)
        cconf2 = float(cconf2)

        conf3 = (cconf2 - cconf)
        cconf3 = '{0:.2f}'.format(conf3)

        janela["vconf"].update(cconf3)

        #IR
        ir = (decimal * 0.015)
        iir = '{0:.2f}'.format(ir)
        iir = float(iir)

        ir2 = (decimal * 0.32)
        iir2 = '{0:.2f}'.format(ir2)
        iir2 = float(iir2)

        ir3 = (iir2 * 0.15)
        iir3 = '{0:.2f}'.format(ir3)
        iir3 = float(iir3)

        ir4 = (iir3 - iir)
        iir4 = '{0:.2f}'.format(ir4)

        janela["vir"].update(iir4)

        #CSLL
        csll = (decimal * 0.01)
        ccsll = '{0:.2f}'.format(csll)
        ccsll = float(ccsll)

        csll2 = (decimal * 0.32)
        ccsll2 = '{0:.2f}'.format(csll2)
        ccsll2 = float(ccsll2)

        csll3 = (ccsll2 * 0.09)
        ccsll3 = '{0:.2f}'.format(csll3)
        ccsll3 = float(ccsll3)

        csll4 = (ccsll3 - ccsll)
        ccsll4 = '{0:.2f}'.format(csll4)

        janela["vcsll"].update(ccsll4)

janela.close()
```
