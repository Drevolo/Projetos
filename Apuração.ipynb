```python
#Automação para realizar Apuração no Sistema Dominio

import pyautogui
import time
import PySimpleGUI as sg

sg.theme('Topanga')

time.sleep(3)
pyautogui.PAUSE = 1

qt = sg.popup_get_text('Varias Empresas ou apenas 1? Responda: (varias) ou (1)')
ap = sg.popup_get_text('Qual o periodo de apuração? Ex: 012024')

Emp = []

if qt == '1':
    em = sg.popup_get_text('Nome da empresa dentro do sistema')
    pyautogui.hotkey('Alt','Tab')
    time.sleep(2)
    pyautogui.click(1814,76) #Mudar de empresa
    pyautogui.write(em)
    pyautogui.press('Down')
    pyautogui.press('Enter')
    time.sleep(5)
    pyautogui.click(379,91) #aba apuração
    time.sleep(2)
    pyautogui.click(835,485) #Colocar o periodo
    pyautogui.write(ap)
    pyautogui.click(1144,442) #iniciar apuração
    time.sleep(10)

    pyautogui.click(1112,501)

    verifica = pyautogui.pixel(1149,490)
    if verifica == (255, 255, 255):
        pyautogui.click(1112,501)
        pyautogui.press('Left')
        pyautogui.press('Enter')
        pass

    else:
        pass

    time.sleep(10)

    def verificar_cor_pixel(posicao, cor, intervalos):
        for intervalo in intervalos:
            if pyautogui.pixelMatchesColor(posicao[0], posicao[1], cor):
                time.sleep(intervalo)
            else:
                realizar_acoes()
                break
        else:
            pass

    def realizar_acoes():
        pyautogui.press('esc')
        pyautogui.press('Enter')
        pyautogui.press('esc')
        

    posicao_pixel = (1019, 377)
    cor_esperada = (157, 184, 212)
    intervalos_espera = [10, 10, 10, 10, 10, 10]

    verificar_cor_pixel(posicao_pixel, cor_esperada, intervalos_espera)

else:
    pass

if qt == 'varias':

    X = sg.popup_get_text('Quantidade')
    M = int(X)

    for i in range(M):

        XX = sg.popup_get_text('Nome das Empresas')
        Emp.append(XX)

    for i in range(M):

        TT = Emp[i]

        time.sleep(2)
        pyautogui.click(1814,76) #Mudar de empresa
        pyautogui.write(TT)
        pyautogui.press('Down')
        pyautogui.press('Enter')
        time.sleep(5)
        pyautogui.click(379,91) #aba apuração
        time.sleep(2)
        pyautogui.click(835,485) #Colocar o periodo
        pyautogui.write(ap)
        pyautogui.click(1144,442) #iniciar apuração
        time.sleep(10)

        pyautogui.click(1112,501)

        verifica = pyautogui.pixel(1149,490)
        if verifica == (255, 255, 255):
            pyautogui.click(1112,501)
            pyautogui.press('Left')
            pyautogui.press('Enter')
            pass

        else:
            pass

        time.sleep(10)

        def verificar_cor_pixel(posicao, cor, intervalos):
            for intervalo in intervalos:
                if pyautogui.pixelMatchesColor(posicao[0], posicao[1], cor):
                    time.sleep(intervalo)
                else:
                    realizar_acoes()
                    break
            else:
                pass

        def realizar_acoes():
            pyautogui.click(1112,501)
            time.sleep(1)
            pyautogui.click(1110,500)
            pyautogui.press('esc')
            pyautogui.press('Enter')
            pyautogui.press('esc')
            

        posicao_pixel = (1019, 377)
        cor_esperada = (157, 184, 212)
        intervalos_espera = [5, 5, 5, 5, 5, 5]

        verificar_cor_pixel(posicao_pixel, cor_esperada, intervalos_espera)

else:
    pass

sg.popup('Finalizado')
```
